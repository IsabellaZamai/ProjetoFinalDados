# 🚗- API de Previsão de Preço de Carros Usados

Uma API REST desenvolvida em **Python** com **Flask** e **Machine Learning** para estimar o preço de carros usados com base em suas características. O projeto utiliza **Scikit-learn** para treinamento do modelo e **Supabase** para armazenar o histórico das previsões realizadas.

---

## 📌 Demonstração

🌐 **Site (Frontend):**
https://auto-worth-guide.lovable.app

🚀 **API Online:**
https://projetofinaldados.onrender.com/

---

# 📖 Sobre o Projeto

O **Auto Worth Guide** foi desenvolvido para auxiliar usuários na estimativa do valor de veículos usados utilizando um modelo de regressão linear treinado com um conjunto de dados contendo informações como:

* Ano do veículo
* Quilometragem
* Potência do motor
* Número de revisões

Após a previsão, os dados da consulta são armazenados automaticamente em um banco de dados no **Supabase**, permitindo manter um histórico das estimativas realizadas.

# 📡 Endpoints

## GET /

Verifica se a API está online.

### Resposta

```json
{
    "resposta": "API online, Flask rodando corretamente",
    "api_principal": "https://projetofinaldados.onrender.com/",
    "site": "https://auto-worth-guide.lovable.app"
}
```

---

## POST /prever

Realiza a previsão do preço do veículo.

### Exemplo de Requisição

```json
{
    "ano": 2020,
    "quilometragem": 45000,
    "motor": 2.0,
    "num_revisoes": 6
}
```

### Exemplo de Resposta

```json
{
    "preco": 87235.42
}
```
